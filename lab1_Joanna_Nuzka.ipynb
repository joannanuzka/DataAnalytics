```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Ex. 2

```python
u = np.random.uniform(0, 1, 1000)
n1 = np.random.normal(0, 1, 1000)
n2 = np.random.normal(0, 10, 1000)
b1 = np.random.beta(1, 1, 1000)
b2 = np.random.beta(2, 2, 1000)
b3 = np.random.beta(3, 9, 1000)
mean = [0, 0]
cov1 = [[1, 0], [0, 1]]
cov2 = [[2, 1], [1, 2]]
n2dim1 = np.random.multivariate_normal(mean, cov1, 10000)
n2dim2 = np.random.multivariate_normal(mean, cov2, 10000)
```

Ex. 3

```python
plt.hist(u, bins = 'auto', density = True)
plt.title('Uniform distribution')
plt.show()
```

```python
plt.hist(n1, bins = 'auto', density = True)
plt.title('Normal distribution, mean = 0, std = 1')
plt.show()
```

```python
plt.hist(n2, bins = 'auto', density = True)
plt.title('Normal distribution, mean = 0, std = 10')
plt.show()
```

```python
plt.hist(b1, bins = 'auto', density = True)
plt.title('Beta distribution, alfa = 1, beta = 1')
plt.show()
```

```python
plt.hist(b2, bins = 'auto', density = True)
plt.title('Beta distribution, alfa = 2, beta = 2')
plt.show()
```

```python
plt.hist(b3, bins = 'auto', density = True)
plt.title('Beta distribution, alfa = 3, beta = 9')
plt.show()
```

The results aren't the same as analytical formulas of distributions but are similar with acceptable tolerancy.

Ex. 4

```python
x1 = n2dim1[:, 0]
y1 = n2dim1[:, 1]
plt.hist2d(x1, y1, density = True, bins=20)
plt.show()
```

```python
x2 = n2dim2[:, 0]
y2 = n2dim2[:, 1]
plt.hist2d(x2, y2, density = True, bins=20)
plt.show()
```

```python
sns.jointplot(x=x1, y=y1)
plt.show()
```

```python
sns.jointplot(x=x2, y=y2)
plt.show()
```

Amount of bins is connected with number of points on plot and with accuracy of visualization. The plots are the most readable for medium number of beans in example 20. Seaborn contains functions which return plot of 2D histogram more readable than this from hist2d method.

Ex. 5

```python
alfa = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
sigma = np.abs(np.random.normal(0, 1, 1000))
mi = alfa + beta * x
y = np.random.normal(mi, sigma)
```

Ex. 6

```python
plt.plot(x, y, 'yo')
xd = np.linspace(-6, 6, 1000)
for i in range(1000):
    yd = alfa[i] + beta[i] * xd
    plt.plot(xd, yd, 'g')
plt.show()
```
